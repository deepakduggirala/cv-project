```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from pathlib import Path
```

```python
data_dir = Path('/Users/deepakduggirala/Documents/Elephants-dataset')
```

```python
data_dir.is_dir()
```

```python
classes = [d for d in data_dir.iterdir() if d.is_dir()]
```

```python
num_class = len(classes)
num_class
```

```python
c = classes[1]
c
```

```python
num_per_class = {c.name:len([i for i in c.iterdir() if not i.name.startswith('.') and (i.name.lower().endswith('jpg'))]) for c in classes}
```

```python
df = pd.DataFrame(num_per_class.items(), columns=['id', 'count'])
```

```python
df2 = df[(df['id']!='1036') & (df['id']!='1038')]
```

```python
df2['count'].sum(), df2['count'].mean(), df2['count'].std(), df2['count'].min(), df2['count'].max()
```

```python
sns.set_theme(style="whitegrid")
sns.barplot(data=df2, x='id', y='count', color='crimson')
plt.xticks(rotation=70)
plt.tight_layout()
plt.title('Images per Elephant')
plt.savefig('Images_per_elephant.png')
```

```python
df2.hist(bins=10)
plt.title('Histogram of images per class')
plt.ylabel('Number of classes')
plt.xlabel('Number of images')
plt.savefig('hist_Images_per_elephant.png')
```

```python
data_dir = Path('/Volumes/Extreme SSD/ELPephant/images')
```

```python
images = sorted([i for i in data_dir.glob('*.jpg') if not i.name.startswith('.')])
```

```python
len(images)
```

```python
from collections import Counter
```

```python
counter = Counter()
```

```python
images[0].stem.split('_')
```

```python
for i in images:
    c = i.stem.split('_')[0]
    counter[c] +=1
```

```python
df = pd.DataFrame(counter.items(), columns=['id', 'count'])
df
```

```python
len(counter.keys())
```

```python
df.hist(bins=10)
plt.title('Histogram of images per class')
plt.ylabel('Number of classes')
plt.xlabel('Number of images')
plt.savefig('hist_Images_per_elephant_ELPephant_dataset.png')
```
