```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from pathlib import Path
```

```python
data_dir = Path('/Users/deepakduggirala/Documents/Elephants-dataset')
classes = [d for d in data_dir.iterdir() if d.is_dir()]
```

```python
def save_as_png(f, png_fname):
    img = Image.open(f, formats=('JPEG',))
    fixed_img = ImageOps.exif_transpose(img)
    fixed_img.save(png_fname, 'PNG')
```

```python
for c in classes:
    for i in c.iterdir():
        print(i)
        target_dir = Path('/Users/deepakduggirala/Documents/Elephants-dataset-png') / c.name 
        target_dir.mkdir(parents=True, exist_ok=True)
        if not i.name.startswith('.') and (i.name.lower().endswith('jpg')):
            target_fname = target_dir / (str(i.stem )+ '.png')
            save_as_png(i, target_fname)
            
```
