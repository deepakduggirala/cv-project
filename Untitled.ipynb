```python
import tensorflow as tf

print(tf.__version__)
```

```python
base_model = tf.keras.applications.InceptionV3(include_top=False, input_shape=(250, 250, 3))
base_model.trainable = False
flatten = tf.keras.layers.Flatten()(base_model.output)
embedding_layer = tf.keras.layers.Dense(units=256)(flatten)
model = tf.keras.Model(base_model.input, embedding_layer, name="model")
```
