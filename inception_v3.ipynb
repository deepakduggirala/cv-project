```python
import tensorflow as tf

print(tf.__version__)

from tensorflow.keras.applications.inception_v3 import preprocess_input
import numpy as np
```

if include_top is True, expected image shape is (299, 299, 3)

width and height should be no smaller than 75

```python
base_model = tf.keras.applications.InceptionV3(
    weights="imagenet",
    include_top=False, 
    input_shape=(299, 299, 3), 
    pooling='avg')
base_model.trainable = False

inputs = tf.keras.Input(shape=(299, 299, 3))
x = base_model(inputs, training=False)
embedding_layer = tf.keras.layers.Dense(units=256)(x)
model = tf.keras.Model(inputs, embedding_layer)
```

```python
for i, layer in enumerate(base_model.layers):
    print(i, layer.name)
```

Total params: 22,327,328

```python
base_model = tf.keras.applications.ResNet50V2(include_top=False,
                                             weights="imagenet",
                                             input_shape=(224,224,3),
                                             pooling='avg')
```

```python
# Freeze all weight till layer conv5_block1_out
trainable = False
for layer in base_model.layers:
    if layer.name == "conv5_block1_out":
        trainable = True
    layer.trainable = trainable
```

```python
trianable = 0
non_trainable = 0
for layer in base_model.layers:
    if layer.trainable:
        trianable += 1
    else:
        non_trainable+=1
        
trianable, non_trainable
```
